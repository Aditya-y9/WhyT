```python
from fpdf import FPDF
import re

def write_back_to_pdf(sections, output_path):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)  # Set the font at the beginning

    for subtopic, content in sections:
        pdf.multi_cell(200, 10, txt=subtopic, align='L')
        pdf.multi_cell(200, 10, txt=content, align='L')
        pdf.ln(5)  # Add some space between sections

    pdf.output(output_path)

def format_text_by_subtopics(text, output_path):
    # Define a regular expression pattern to identify subtopics and their content
    subtopic_pattern = re.compile(r'(\w[\w\s]*):\s*(.*?)(?=\n\w+:|$)', re.DOTALL)

    # Find all occurrences of subtopics in the text
    matches = subtopic_pattern.findall(text)

    # Print matched sections for debugging
    print("Matched Sections:")
    print(matches)

    write_back_to_pdf(matches, output_path)

# Example usage
pdf_path = r"C:\Users\MSHOME\Desktop\Newfolder\temp\inheritance-project\transcript_output.pdf"
output_path = "formatted_output.pdf"

text = extract_text_from_pdf(pdf_path)
print("Extracted Text:")
print(text)

format_text_by_subtopics(text, output_path)
```

```python
from fpdf import FPDF
import re

def write_back_to_pdf(paragraphs, output_path):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)  # Set the font at the beginning

    for paragraph in paragraphs:
        pdf.multi_cell(200, 10, txt=paragraph, align='L')
        pdf.ln(5)  # Add some space between paragraphs

    pdf.output(output_path)

def format_text_by_paragraphs(text, output_path):
    # Split the text into paragraphs
    paragraphs = text.split('\n')

    # Print paragraphs for debugging
    print("Paragraphs:")
    print(paragraphs)

    write_back_to_pdf(paragraphs, output_path)

# Example usage
pdf_path = r"C:\Users\MSHOME\Desktop\Newfolder\temp\inheritance-project\transcript_output.pdf"
output_path = "formatted_output.pdf"

text = extract_text_from_pdf(pdf_path)
print("Extracted Text:")
print(text)

format_text_by_paragraphs(text, output_path)
```
